## Função de Distribuição Acumulada (*Cumulative Distribution Function*)

A distribuição cumulativa pode ser interpretada como: 

> “Qual é a chance de um valor selecionado aleatoriamente ser menor do que X?”. 

Como a probabilidade sempre aumenta ou permanece constante com valores mais altos de $x$, a cdf é uma função monotônica.

Existe apenas um parâmetro, $\lambda$, usado para diferenciar entre distribuições exponenciais que governam a curvatura da cdf.

$$
\text{F}(x) = 1 - e^{-\lambda x}
$$

```python
def cdf(lam, x):
    return 1 - np.exp(-lam * x)

def plot_cdf(lam, y_scale=1):
    x = np.linspace(0,1,100)
    y = [cdf(lam, xi) * y_scale for xi in x]
    plt.plot(x, y, label="$\lambda=%g$" % lam, lw=2)

[plot_cdf(lam) for lam in [8.6, 3.6, 1.6, 0.1]]
plt.xlabel("x")
plt.ylabel("$Quantile\ F(x)$")
plt.legend(loc='lower right')
plt.grid()
```

## Função de Densidade de Probabilidade (*Probability Density Function*)

$$
\text{pdf}(x) = \frac{d}{dx} F(x) = \lambda e^{-\lambda x}
$$

```python
def pdf(lam, x):
    return lam * np.exp(-lam * x)

def plot_pdf(lam):
    x = np.linspace(0,1,100)
    y = [pdf(lam, xi) for xi in x]
    plt.plot(x, y, label="$\lambda=%g$" % lam, lw=2)

[plot_pdf(lam) for lam in [8.6, 3.6, 1.6, 0.1]]
plt.xlabel("x")
plt.ylabel("$pdf(x)$")
plt.legend()
plt.grid()
```
