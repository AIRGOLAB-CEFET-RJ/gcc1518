# Distribuição de Bernoulli

Simula o lançamento de uma moeda com probabilidade p de cara.

```python
from scipy.stats import bernoulli

# Probabilidade de cara
p = 0.6

# Lançar a moeda 10 vezes
lancamentos = bernoulli.rvs(p, size=10)

# Probabilidade de cara em cada lançamento
print(lancamentos)
```

# Distribuição Uniforme

Gera números aleatórios discretos entre $a$ e $b$ com probabilidade igual.

```python
from scipy.stats import uniform

# Limites inferior e superior
a = 5
b = 15

# Gerar 10 números aleatórios
numeros = uniform.rvs(a, b-a, size=10)

# Mostrar os números gerados
print(numeros)
```

# Distribuição Binomial

Calcula a probabilidade de $k$ sucessos em $n$ tentativas em um experimento binomial com probabilidade de sucesso $p$.

```python
from scipy.stats import binom

# Número de tentativas
n = 10

# Probabilidade de sucesso
p = 0.4

# Número de sucessos desejado
k = 3

# Probabilidade de obter k sucessos
probabilidade = binom.pmf(k, n, p)

print(f"Probabilidade de {k} sucessos em {n} tentativas: {probabilidade}")
```

# Distribuição Geométrica

Calcula a probabilidade de que o $k$-ésimo evento seja um sucesso em uma série de experimentos independentes com probabilidade de sucesso $p$.

```python
from scipy.stats import geom

# Probabilidade de sucesso
p = 0.2

# Evento desejado (k-ésimo sucesso)
k = 5

# Probabilidade do k-ésimo sucesso
probabilidade = geom.pmf(k-1, p)

print(f"Probabilidade do k-ésimo sucesso: {probabilidade}")
```

# Distribuição Binomial Negativa

Calcula a probabilidade de precisar de $k$ tentativas para obter $r$ sucessos em um experimento binomial com probabilidade de sucesso $p$.

```python
from scipy.stats import nbinom

# Número de sucessos desejado
r = 2

# Probabilidade de sucesso
p = 0.3

# Número de tentativas necessário
k = 5

# Probabilidade de precisar de k tentativas para r sucessos
probabilidade = nbinom.pmf(k-1, r, p)

print(f"Probabilidade de precisar de {k} tentativas para {r} sucessos: {probabilidade}")
```

# Distribuição Multinomial

Calcula a probabilidade de obter k resultados específicos em n tentativas em um experimento multinomial com probabilidades p.

```python
from scipy.stats import multinomial

# Número de tentativas
n = 15

# Probabilidades de cada resultado
probabilidades = [0.2, 0.5, 0.3]

# Número de resultados específicos
k = [3, 6, 6]

# Probabilidade de obter k resultados específicos
probabilidade = multinomial.pmf(k, n, probabilidades)

print(f"Probabilidade de obter {k} resultados: {probabilidade}")
```

# Distribuição Hipergeométrica

Calcula a probabilidade de obter k resultados específicos em uma amostra n de tamanho total N com K resultados do tipo específico na população.

Imagine que temos um conjunto de $N=20$ animais, dos quais $k=7$ são cachorros. Se quisermos saber a probabilidade de encontrar um determinado número de cachorros ao escolhermos aleatoriamente $n=12$ dos 20 animais, podemos inicializar uma distribuição congelada e plotar a função de massa de probabilidade.

```python
import numpy as np
from scipy.stats import hypergeom
import matplotlib.pyplot as plt
```

```python
import numpy as np

[N, k, n] = [20, 7, 12]
rv = hypergeom(N, k, n)
x = np.arange(0, k+1)
pmf_dogs = rv.pmf(x)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, pmf_dogs, 'bo')
ax.vlines(x, 0, pmf_dogs, lw=2)
ax.set_xlabel('# of dogs in our group of chosen animals')
ax.set_ylabel('hypergeom PMF')
plt.show()
```

# Distribuição de Poisson

Calcula a probabilidade de $x$ eventos em um intervalo de tempo dado com taxa média de eventos $\lambda$.

```python
from scipy.stats import poisson

# Taxa média de eventos (por exemplo, eventos por hora)
lambda_ = 3

# Intervalo de tempo (por exemplo, horas)
tempo = 5

# Calcular probabilidades para cada número possível de eventos
eventos = range(tempo + 1)
probabilidades = poisson.pmf(eventos, lambda_*tempo)

# Visualizar a função de massa de probabilidade (PMF)
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.bar(eventos, probabilidades, color='skyblue', edgecolor='k')
plt.title(f"Distribuição de Poisson (λ = {lambda_}, tempo = {tempo})")
plt.xlabel("Número de eventos")
plt.ylabel("Probabilidade")
plt.grid(True)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(eventos, probabilidades, 'bo')
ax.vlines(eventos, 0, probabilidades, lw=2)
plt.title(f"Distribuição de Poisson (λ = {lambda_}, tempo = {tempo})")
plt.xlabel("Número de eventos")
plt.ylabel("Probabilidade")
plt.show()
```
