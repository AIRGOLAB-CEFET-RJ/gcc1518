```python
import numpy as np
import matplotlib.pyplot as plt

class ThompsonSampling:
    def __init__(self, num_arms):
        self.num_arms = num_arms
        self.alpha = np.ones(num_arms)  # Initialize alpha parameters to 1
        self.beta = np.ones(num_arms)   # Initialize beta parameters to 1
        
    def select_arm(self):
        sampled_means = np.random.beta(self.alpha, self.beta)
        return np.argmax(sampled_means)
    
    def update(self, arm, reward):
        if reward == 1:
            self.alpha[arm] += 1
        else:
            self.beta[arm] += 1
            
    def plot_arm_distribution(self):
        x = np.linspace(0, 1, 100)
        for arm in range(self.num_arms):
            y = np.random.beta(self.alpha[arm], self.beta[arm], 1000)
            plt.hist(y, bins=30, density=True, alpha=0.5, label=f'Arm {arm+1}')
        plt.title('Beta Distribution of Arms')
        plt.xlabel('Probability')
        plt.ylabel('Density')
        plt.legend()
        plt.show()

# Example usage
num_arms = 5
bandit = ThompsonSampling(num_arms)

# Simulate interactions
num_iterations = 1000
rewards = np.zeros(num_iterations)
for i in range(num_iterations):
    arm = bandit.select_arm()
    # Simulate reward, for example, 1 for success, 0 for failure
    reward = np.random.choice([0, 1], p=[0.8, 0.2])  # Adjust probabilities as needed
    bandit.update(arm, reward)
    rewards[i] = reward

# Plot arm distributions
bandit.plot_arm_distribution()
```
