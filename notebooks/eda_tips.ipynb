O código abaixo produz um gráfico de dispersão entre as variáveis `total_bill` e `tip` e adiciona uma linha de regressão a este gráfico. Para isso, usamos a função `lmplot` do Seaborn. Essa função facilita a criação de gráficos de dispersão com linhas de regressão ajustadas.

No gráfico produzido, os pontos abaixo da linha correspondem a gorjetas inferiores ao esperado (para aquele valor da conta) e os pontos acima da linha são superiores ao esperado. Poderíamos esperar ver uma associação linear positiva e estreita, mas em vez disso veremos uma variação que aumenta com o valor da gorjeta. Em particular, há mais pontos distantes da linha no canto inferior direito do que no canto superior esquerdo, indicando que mais clientes são muito baratos do que muito generosos.

```python
# 1d

import seaborn as sns
import matplotlib.pyplot as plt

# Carregar o dataset 'tips'
tips = sns.load_dataset('tips')

# Criar o gráfico de dispersão com linha de regressão
plt.figure(figsize=(10, 6))
sns.lmplot(x='total_bill', y='tip', data=tips, aspect=1.5, line_kws={'color': 'red'}, ci=0)
plt.title('Gráfico de Dispersão com Linha de Regressão: Total da Conta vs. Gorjeta')
plt.xlabel('Total da Conta ($)')
plt.ylabel('Gorjeta ($)')
plt.show()
```

- As festas para fumantes têm muito mais variabilidade nas gorjetas dadas. 
- Os homens tendem a pagar as (poucas) contas mais altas.
- As mulheres não fumantes tendem a dar gorjetas cujos valores são muito consistentes (com três exceções visíveis mostradas na amostra).

```python
# 1d

import seaborn as sns
import matplotlib.pyplot as plt

# Carregar o dataset 'tips'
tips = sns.load_dataset('tips')

# Criar um FacetGrid para múltiplos subplots
g = sns.FacetGrid(tips, col="sex", row="smoker", margin_titles=True, height=4, aspect=1.2)
g.map_dataframe(sns.regplot, x="total_bill", y="tip")

# Adicionar títulos e rótulos
g.set_axis_labels('Total da Conta ($)', 'Gorjeta ($)')
g.set_titles(col_template="{col_name}", row_template="{row_name}")
plt.subplots_adjust(top=0.9)
g.fig.suptitle('Scatterplot de Gorjetas vs. Total da Conta\npor Gênero e Status de Fumante')

plt.show()
```

O código abaixo produz um histograma dos valores de gorjeta do conjunto de dados tips utilizando o Seaborn, onde as caixas (bins) cobrem incrementos de $1. Aqui, usamos a função `histplot` com uma configuração apropriada para os bins.

Repare que distribuição de valores é distorcida para a direita e unimodal, como é comum em distribuições de quantidades pequenas e não negativas.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar o conjunto de dados 'tips'
tips = sns.load_dataset('tips')

# Criar um histograma dos valores de gorjeta com bins cobrindo incrementos de $1
plt.figure(figsize=(10, 6))
sns.histplot(data=tips, x='tip', bins=range(0, int(tips['tip'].max()) + 2, 1), color='skyblue', edgecolor='black')
plt.title('Histograma dos Valores de Gorjeta')
plt.xlabel('Valor da Gorjeta ($)')
plt.ylabel('Frequência')
plt.xticks(range(0, int(tips['tip'].max()) + 2, 1))  # Definir os ticks do eixo x para coincidir com os limites dos bins
plt.show()
```

Podemos produzir outro histograma, desta vez alterando o bin para $0.1. Um fenômeno interessante se torna visível: os picos ocorrem nos valores de dólar inteiro e meio dólar, o que é causado pelo fato de os clientes escolherem números redondos como gorjetas. Esse comportamento também é comum em outros tipos de compras, como a gasolina.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Carregar o dataset 'tips'
tips = sns.load_dataset('tips')

# Determinar o intervalo para os bins de $0.1
bin_width = 0.1
bins = np.arange(0, tips['tip'].max() + bin_width, bin_width)

# Criar o histograma dos valores de gorjeta com bins de $0.1
plt.figure(figsize=(10, 6))
sns.histplot(data=tips, x='tip', bins=bins, color='skyblue', edgecolor='black')
plt.title('Histograma dos Valores de Gorjeta com Incrementos de $0.1')
plt.xlabel('Valor da Gorjeta ($)')
plt.ylabel('Frequência')
plt.show()
```

```python
# 3a. Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=tips, x='day', y='total_bill', palette="Set2")
plt.title('Boxplot do Total da Conta por Dia da Semana')
plt.xlabel('Dia da Semana')
plt.ylabel('Total da Conta ($)')
plt.show()
```

```python
# 4. Gráfico de Violino
plt.figure(figsize=(10, 6))
sns.violinplot(data=tips, x='day', y='tip', hue='time', split=True, palette="muted")
plt.title('Gráfico de Violino da Gorjeta por Dia da Semana e Hora')
plt.xlabel('Dia da Semana')
plt.ylabel('Gorjeta ($)')
plt.show()
```

```python
# 5. Gráfico de Barras
plt.figure(figsize=(10, 6))
sns.barplot(data=tips, x='day', y='total_bill', ci=None, palette='coolwarm')
plt.title('Gráfico de Barras do Total da Conta por Dia da Semana')
plt.xlabel('Dia da Semana')
plt.ylabel('Total da Conta ($)')
plt.show()
```

```python
# 6a. Gráfico de Calor (Heatmap) da Tabela de Contingência
plt.figure(figsize=(10, 6))
pivot_table = tips.pivot_table(index='day', columns='time', values='total_bill', aggfunc='mean')
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='.2f')
plt.title('Heatmap da Média do Total da Conta por Dia e Hora')
plt.xlabel('Hora do Dia')
plt.ylabel('Dia da Semana')
plt.show()
```
